#### 1. Creating Tensors

```python
## 1.1 Scalar ( a single number)
scalar = torch.tensor(7)
scalar
```

```python
# let's print more about it
print(f"Scalar : {scalar}")
print(f"Dimension : {scalar.ndim}")
print(f"Shape : {scalar.shape}")
print(f"Device on which scalar is running : {scalar.device}")
```

```python
# 1.2 Vector (a 1-D array of numbers)
vector = torch.tensor([1,2,3])
```

```python
print(f"Vector : {vector}")
print(f"Dimension : {vector.ndim}")
print(f"Shape : {vector.shape}")  # (n,) n is number of elements
print(f"Device on which vector is running : {vector.device}")
```

```python
# 1.3 Matrix (a 2-D array of numbers)
MATRIX = torch.tensor([[1, 2 ,3],
                       [5, 6 , 7]
])
```

```python
print(f"Matrix : {MATRIX}")
print(f"Dimension : {MATRIX.ndim}")
print(f"Shape : {MATRIX.shape}")  # (rows , columns)
```

```python
# 1.4 Tensor (a generalization of all to n dimensions)
TENSOR= torch.tensor([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
```

```python
print(f"Tensor : {TENSOR}")
print(f"Dimension : {TENSOR.ndim}")
print(f"Shape : {TENSOR.shape}")
```

Things to remember:

A Tensor is a `genralization` of scalar , vector and matrix to `n` dimensions

* A Tensor of `0` dimension is a `Scalar`.
* A Tensor of `1` dimension is a `Vector`.
* A Tensor of `2` dimension is a `Matrix`.

Generally we called Tensor to a Tensor have more than `2` dimension.

Note:

* *`torch.Size` (or .shape) tells you how many elements exist along each dimension of a tensor.*
* For a tensor with shape [d0, d1, d2, ..., dn]:

     Number of values in the shape = number of dimensions of the tensor.

     Each value = size (length) of that specific dimension.
     
* Total number of elements = product of all shape values


```python
# 1.5 Zero Tensor (having 0's as an  element)
zero_tensor = torch.zeros(size = (2,2))
print(f"Tensor : {zero_tensor}")
print(f"Dimension : {zero_tensor.ndim}")
print(f"Shape : {zero_tensor.shape}")
```

```python
# 1.6 Ones Tensor (having 1's as an element)
ones_tensor = torch.ones(size = (3,3))
print(f"Tensor : {ones_tensor}")
print(f"Dimension : {ones_tensor.ndim}")
print(f"Shape : {ones_tensor.shape}")
```

```python
# 1.7 Range Tensor
range_tensor = torch.arange(start = 1.5,
                            step = 1,
                            end = 10.98)
range_tensor
```

Note: In Python, you can use `range()` to create a range. However in PyTorch, `torch.range()` is deprecated and may show an error in the future.

```python
# 1.8 Random tensor
random_tensor = torch.rand(size = (2,3))
random_tensor
```

`torch.rand()` Returns a tensor filled with random numbers from a uniform distribution on the interval [0,1).

The shape of the tensor is defined by the variable argument `size`.

Note:

In PyTorch, both `torch.rand()` and `torch.randn()` generate random tensors, but they differ in the distribution of the generated values.

`torch.rand()` Generates numbers uniformly between 0 and 1

`torch.randn()` Generates numbers from a standard normal distribution (mean = 0, std = 1).

```python
random = torch.randn(size = (2,3))
random
```

```python
# Let's check it's mean and standard deviation (it will be nearly 0 and 1 not exact!)
mean = random.mean()
std = random.std()

print(f"Mean is {mean}")
print(f"Standard deviation is {std}")
```
