Things to remember:

A Tensor is a `genralization` of scalar , vector and matrix to `n` dimensions

* A Tensor of `0` dimension is a `Scalar`.
* A Tensor of `1` dimension is a `Vector`.
* A Tensor of `2` dimension is a `Matrix`.

Generally we called Tensor to a Tensor have more than `2` dimension.

Note:

* *`torch.Size` (or .shape) tells you how many elements exist along each dimension of a tensor.*
* For a tensor with shape [d0, d1, d2, ..., dn]:

     Number of values in the shape = number of dimensions of the tensor.

     Each value = size (length) of that specific dimension.
     
* Total number of elements = product of all shape values

`torch.rand()` Returns a tensor filled with random numbers from a uniform distribution on the interval [0,1).

The shape of the tensor is defined by the variable argument `size`.

Note:

In PyTorch, both `torch.rand()` and `torch.randn()` generate random tensors, but they differ in the distribution of the generated values.

`torch.rand()` Generates numbers uniformly between 0 and 1

`torch.randn()` Generates numbers from a standard normal distribution (mean = 0, std = 1).

```python
random = torch.randn(size = (2,3))
random
```

```python
# Let's check it's mean and standard deviation (it will be nearly 0 and 1 not exact!)
mean = random.mean()
std = random.std()

print(f"Mean is {mean}")
print(f"Standard deviation is {std}")
```

##### Matrix Multiplication (backbone of deep learning!)

One of the most common operations in machine learning and deep learning algorithms (like neural networks) is matrix multiplication.

PyTorch implements matrix multiplication functionality in the `torch.matmul()` method.


The main two rules for matrix multiplication to remember are:

1. The inner dimensions must match:

* (3, 2) @ (3, 2) won't work

* (2, 3) @ (3, 2) will work

* (3, 2) @ (2, 3) will work

2. The resulting matrix has the shape of the outer dimensions:

* (2, 3) @ (3, 2) -> (2, 2)

* (3, 2) @ (2, 3) -> (3, 3)

Note: "@" in Python is the symbol for matrix multiplication.

```python
# Create tensors
X = torch.randn(size = (2,3))
Y = torch.randn(size = (3,4))

X.shape , Y.shape
```

Both the matrices are compatible for matrix multiplication as their inner dimensions are same

```python
print(torch.matmul(X,Y))
print(torch.matmul(X,Y).shape)
```

##### One of the most common errors in deep learning (shape errors)

Because much of deep learning is multiplying and performing operations on matrices and matrices have a strict rule about what shapes and sizes can be combined, one of the most common errors you'll run into in deep learning is shape mismatches.

```python
# Shapes need to be in the right way  
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)

torch.matmul(tensor_A, tensor_B) # (this will error)
```

We can make matrix multiplication work between tensor_A and tensor_B by making their inner dimensions match.

One of the ways to do this is with a transpose (switch the dimensions of a given tensor).

You can perform transposes in PyTorch using either:

`torch.transpose(input, dim0, dim1)` - where input is the desired tensor to transpose and dim0 and dim1 are the dimensions to be swapped.

`tensor.T` - where tensor is the desired tensor to transpose.

```python
print(tensor_A)
print(tensor_B.T)
```

```python
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output) 
print(f"\nOutput shape: {output.shape}")
```

#### 4. Finding the min, max, mean, sum, etc (aggregation)

```python
# create a tensor
x = torch.arange(0,100,10)
x
```

```python
# Now perform some aggregation
print(f"Minimum: {x.min()}")
print(f"Maximum: {x.max()}")
print(f"Sum: {x.sum()}")
# print(f"Mean: {x.mean()}") # this will error
print(f"Mean: {x.type(torch.float32).mean()}") # won't work without float datatype
```

Note: You may find some methods such as `torch.mean()` require tensors to be in `torch.float32` (the most common) or another specific datatype, otherwise the operation will fail.

```python
# We can also do above with torch 
torch.max(x), torch.min(x), torch.mean(x.type(torch.float32)), torch.sum(x)
```

#### 7. Indexing (selecting data from tensors)

```python
# create a tensor
x = torch.arange(1,10).reshape(1,3,3)
x , x.shape
```

 General form of the index `x[batch, row, column]`

```python
# Select first batch of x
x[0]
```

```python
# select rows from the 1st batch
print(f"First row : {x[0,0]}")
print(f"second row : {x[0,1]}")
print(f"Third row : {x[0,2]}")
```

```python
# Slect a single element from a specific row from the the specific batch
print(f"First element of the first row of the first batch is : {x[0,0,0]}")
print(f"Second element of the second row of the first batch is : {x[0,1,1]}")
```

The colon `:` operator

A colon means `select everything from this dimension`.

```python
# select all batches row 1
print(x[:,1,:])
print(x[:,1,:].shape)
```

```python
# Select all rows , column 1
x[:,:,1]
```

```python
# select the first column
x[:,:,0]
```
